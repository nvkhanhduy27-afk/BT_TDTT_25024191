# W3A1
Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình “SỐ 
NGUYÊN” đảo ngược của n. Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử 
lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý. 

```python
n = int(input("Nhập số nguyên dương có 3 chữ số: "))
if 100 <= n <= 999:
    units = n % 10
    tens = (n // 10) % 10
    hundreds = n // 100
    reversed_num = units * 100 + tens * 10 + hundreds
    print(f"Số nguyên đảo ngược của {n} là: {reversed_num}")
else:
    print("Số nhập vào không phải là số nguyên dương có 3 chữ số.")
```

# W3A3
Kiểm tra xem một số có phải là lũy thừa của 2 hay không 
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 
hay không. 

```python
n = int(input("Nhập n"))
print(n)
if n > 0 and (n & (n - 1)) == 0:
    print("là lũy thừa của 2.")
else:
    print("Số vừa nhập không phải là lũy thừa của 2.")
```

# W3A4

# W3A16
[Rounding] 
Nhập vào một số thực: in ra kết quả làm tròn lên, tròn xuống, làm tròn tới số 
nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả 
cách nhau một dấu cách

```python
x = float(input("Nhập số thực: "))
tron_xuong = int(x) if x >= 0 or x == int(x) else int(x) - 1
tron_len = int(x) if x <= 0 or x == int(x) else int(x) + 1
phan_thap_phan = x - tron_xuong
if phan_thap_phan >= 0.5:
    gan_nhat = tron_xuong + 1
else:
    gan_nhat = tron_xuong
print(f"{int(tron_len)} {int(tron_xuong)} {int(gan_nhat)}")
```

# W3A17
Cấp số nhân  
Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 
cấp số nhân với công bội nguyên theo đúng thứ tự a, b, c, d hay không?  
Gợi ý : Tìm công bội (b / a) nếu b chia hết cho a, sau đó lấy b nhân công bội 
và so sánh vs c, c nhân công bội và so sánh vs d. 

```python
a = int(input())
b = int(input())
c = int(input())
d = int(input())
print(a, b, c, d)
if a != 0 and b % a == 0:
    q = b // a 
    if b * q == c and c * q == d:
        print("true")
    else:
        print("false")
else:
    if a == b == c == d == 0:
        print("true")
    else:
        print("false")
```
