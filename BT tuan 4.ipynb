# W4A1
Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để 
tính tống các số nguyên từ 1 đến n. (n <= 1000) 

```python
n = int(input("Nhập n (<= 1000):"))
print(n)
if 1 <= n <= 1000:
    print("n trong [1, 1000]")
    tong = 0
    for i in range(1, n + 1):
        tong += i
    print("Tổng từ 1 đến n là:", tong)
else:
    print("n không hợp lệ")
```

# W4A2
VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi. 
Kiểm tra số này có phải là số nguyên tố hay không?

```python
import math
while True:
    try:
        n = int(input("Nhập một số nguyên dương: "))
        if n > 0 :
            break
        else:
            print("Nhập một số lớn hơn 0")
    except ValueError:
        print("Nhập một số nguyên hợp lệ")
is_prime = True
if n < 2:
    is_prime = False
else:
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            is_prime = False
            break
if is_prime:
    print(f"Số {n} là số nguyên tố")
else:
    print(f"Số {n} không là số nguyên tố")
```

# W4A4

của số đó. Không dùng kiểu dữ liệu string.

```python
n = int(input("Nhập vào một số nguyên n"))
temp = abs(n)
count = 0
if temp == 0:
    while temp > 0:
        count == 1
else:
    while temp > 0:
        temp //= 10
        count += 1
print(f"SỐ chữ số của {n} là :{count}")
```

# W4A6 
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn [a, b]. Mỗi dòng là một bộ test (a b)." 

```python
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0: return False
    return True
import sys
for line in sys.stdin:
    if not line.strip(): break
    a, b = map(int, line.split())
    tong = 0
    for i in range(a, b + 1):
        if is_prime(i):
            tong += i
    print(tong)
```

# W4A8
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số 
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
n = int(input())
steps = 0 
while True:
    s = str(n)
    if s == s[:: -1]:
        print(f"{steps}{n}")
        break
    n = n + int(s[:: -1])
    steps += 1
```

# W4A9 
Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu cách). Nếu không có, in ra (no number).

```python
n = int(input())
x = []
i = 1
while i * i <= n:
    s = str(i * i)
    if len(s) == len(set(s)):
        x.append(s)
    i += 1
if x:
    print(" ".join(x))
else:
    print("no number")
```

# W4A10
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất 
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số 
nhỏ nhất."

```python
n = int(input())
max_length = 0
best_x = 0
for i in range(1, n + 1):
    current_x = i
    length = 1
    temp = current_x
    while temp > 1:
        if temp % 2 == 0:
            temp //= 2
        else:
            temp = 3 * temp + 1
        length += 1
    if length > max_length:
        max_length = length
        best_x = current_x
print(best_x, max_length)
```

# W4A13
Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật: 
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại. 
Ví dụ 220 và 284 là một cặp số thân thiết 
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55, 
110. Tổng của 11 ước này cho chúng ta 284 
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng 
vừa đúng bằng 220 
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. 
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
def s(n):
    x = 1
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            x += i
            if i * i != n: x += n // i
    return x
try:
    a, b = map(int, input().split())
    print(a, b)
    print("true" if s(a) == b and s(b == a) else "false")
except:
    pass
```

```python
n, m = map(int, input().split())
print(n, m)
cho = (m - 2 * n) / 2
ga = n - cho
if m % 2 == 0 and 0 <= ga <= n and cho == int(cho):
    print(int(ga), int(cho))
else:
    print("invalid")
```
